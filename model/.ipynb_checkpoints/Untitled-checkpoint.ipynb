```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 10)
pd.set_option('max_colwidth', 1600)
import re
from sklearn import linear_model
from scipy import stats
import mpld3
mpld3.enable_notebook()
```

```python
LINEUP_THRESHOLD = 24 * 60
YEAR = '201718'
```

```python
all_matchups = pd.read_csv('%s.csv' % YEAR, names=['home_lineup', 'away_lineup', 'home_score', 'away_score', 'time'])
all_matchups = all_matchups[all_matchups.time >= 24]

nba_lineups_traditional = pd.read_csv('nba_lineups_traditional_%s.csv' % YEAR, usecols=['GROUP_NAME', 'TEAM_ABBREVIATION', 'GP', 'MIN', 'PTS', 'FG_PCT', 'PLUS_MINUS'])
nba_lineups_traditional['PPM'] = nba_lineups_traditional.apply(lambda x: x['PTS'] / x['MIN'], axis=1)

nba_lineups_advanced = pd.read_csv('nba_lineups_advanced_%s.csv' % YEAR, usecols=['GROUP_NAME', 'TEAM_ABBREVIATION', 'GP', 'W', 'L', 'W_PCT', 'MIN', 'OFF_RATING', 'DEF_RATING', 'NET_RATING', 'EFG_PCT', 'AST_PCT'])

player_advanced = pd.read_csv('player_advanced_%s.csv' % YEAR, usecols=['PLAYER_NAME', 'NET_RATING'])
```

```python
lineup_dict = {}
num_matchups = len(all_matchups.index)
for i in range(0, num_matchups):
    home = all_matchups['home_lineup'].values[i]
    away = all_matchups['away_lineup'].values[i]
    time = all_matchups['time'].values[i]
    if home not in lineup_dict:
        lineup_dict[home] = time
    else:
        lineup_dict[home] += time
    if away not in lineup_dict:
        lineup_dict[away] = time
    else:
        lineup_dict[away] += time
valid_lineups_by_time = dict((key, value) for key, value in lineup_dict.iteritems() if value >= LINEUP_THRESHOLD)
lineup_counter = 0
valid_lineup_dict = {}
for k, v in valid_lineups_by_time.iteritems():
    if k not in valid_lineup_dict:
        valid_lineup_dict[k] = lineup_counter
        lineup_counter += 1
valid_lineup_lookup = dict((v, k) for (k, v) in valid_lineup_dict.iteritems())
valid_matchups_lookup = all_matchups.isin({'home_lineup': valid_lineups_by_time.keys(), 'away_lineup': valid_lineups_by_time.keys()})
valid_matchups_lookup['valid'] = valid_matchups_lookup.apply(lambda x: (x['home_lineup'] and x['away_lineup']), axis = 1)
valid_matchups = all_matchups[valid_matchups_lookup['valid'] == True]
```

```python
num_lineups = len(valid_lineup_dict)
num_matchups = len(valid_matchups.index)
print("%d lineups, %d matchups"% (num_lineups, num_matchups))
```

```python
A = np.zeros((num_matchups, num_lineups + 1))
b = np.zeros(num_matchups)
i = 0
for idx, val in valid_matchups.iterrows():
    home_id = valid_lineup_dict[val['home_lineup']]
    away_id = valid_lineup_dict[val['away_lineup']]
    diff = val['home_score'] - val['away_score']
    time = val['time']
    A[i][home_id] = time
    A[i][away_id] = -1 * time
    A[i][num_lineups] = time
    b[i] = diff
    i += 1
```

```python
def get_model(A, y, l = 0.5):
    reg = linear_model.BayesianRidge(tol=1.e-4, compute_score=True, fit_intercept=False, n_iter=1000, verbose=True)
    reg.fit(A, y)
    return reg.coef_
```

```python
t = get_model(A, b, 10)
mu = np.mean(t)
sd = np.var(t) ** 0.5
plt.hist(t, bins='auto', normed=True)
plt.plot(sorted(t), stats.norm.pdf(sorted(t), mu, sd))
plt.show()
```

```python
home_court_advantage = t[-1]
lineup_strengths = t[0:-1]
print("Home court advantage: %f points per 48" % (home_court_advantage * 48 * 60))
```

```python
lines = pd.DataFrame(data={'id': range(0, num_lineups)})
lines['lineup'] = lines.apply(lambda x: valid_lineup_lookup[x['id']], axis=1)
lines['strength'] = lines.apply(lambda x: lineup_strengths[x['id']], axis=1)
lines['minutes'] = lines.apply(lambda x: valid_lineups_by_time[x['lineup']] / 60, axis=1)
```

```python
lines.sort_values('strength', ascending=False)
```

```python
print("Difference between best and worst teams: %f points per 48" % ((lines.sort_values(by='strength', ascending=False)['strength'].values[0] - lines.sort_values(by='strength', ascending=False)['strength'].values[1]) * 48 * 60))
```

```python
player_lineups = {}
for i in range(0, num_lineups):
    lineup = valid_lineup_lookup[i]
    lineup_players = lineup.split("+")
    for j in range(0, len(lineup_players)):
        player = lineup_players[j]
        if player not in players:
            player_lineups[player] = [{'id': i, 'mins': lines[lines['id'] == i]['minutes'].values[0], 'strength': lines[lines['id'] == i]['strength'].values[0]}]
        else:
            player_lineups[player] += [{'id': i, 'mins': lines[lines['id'] == i]['minutes'].values[0], 'strength': lines[lines['id'] == i]['strength'].values[0]}]
player_dict = {}
num_players = 0
for k, v in player_lineups.iteritems():
    player_dict[k] = num_players
    num_players += 1
player_lookup = dict((v, k) for (k, v) in player_dict.iteritems())
```

```python
player_list = []
for player, data in player_lineups.iteritems():
    mins_sum = sum(map(lambda x: x['mins'], data))
    sqrt_mins_sum = sum(map(lambda x: x['mins'] ** 0.5, data))
    weighted_strength_sum = sum(map(lambda x: (x['mins']) * x['strength'], data))
    weighted_avg_strength = weighted_strength_sum / (1. * mins_sum)
    sqrt_weighted_strength_sum = sum(map(lambda x: (x['mins'] ** 0.5) * x['strength'], data))
    sqrt_weighted_avg_strength = sqrt_weighted_strength_sum / (1. * sqrt_mins_sum)
    player_list.append([player, weighted_avg_strength, mins_sum, map(lambda x: x['id'], data), map(lambda x: x['mins'], data)])
```

```python
player_df = pd.DataFrame(player_list, columns=['name','strength','minutes','lineups', 'lineup_minutes'])
player_df = player_df[player_df['minutes'] > 500]
```

```python
player_df.sort_values('strength', ascending=False)
```

```python
def findMatchingLineupAdvanced(lineup, toPrint=False):
    players = lineup.split("+")    
    player_first_names = map(lambda name: re.search(r"^[A-Z][a-z]*", name).group(), players)
    player_last_names = map(lambda name: re.search(r"[A-Z][a-z]*$", name).group(), players)
    if toPrint == True:
        print player_first_names
        print player_last_names
    return nba_lineups_advanced[nba_lineups_advanced['GROUP_NAME'].str.contains(player_first_names[0]) & nba_lineups_advanced['GROUP_NAME'].str.contains(player_first_names[1]) & nba_lineups_advanced['GROUP_NAME'].str.contains(player_first_names[2]) & nba_lineups_advanced['GROUP_NAME'].str.contains(player_first_names[3]) & nba_lineups_advanced['GROUP_NAME'].str.contains(player_first_names[4])  & nba_lineups_advanced['GROUP_NAME'].str.contains(player_last_names[0]) & nba_lineups_advanced['GROUP_NAME'].str.contains(player_last_names[1]) & nba_lineups_advanced['GROUP_NAME'].str.contains(player_last_names[2]) & nba_lineups_advanced['GROUP_NAME'].str.contains(player_last_names[3]) & nba_lineups_advanced['GROUP_NAME'].str.contains(player_last_names[4])]
```

```python
def findMatchingPlayerAdvanced(name):
    first_name = re.search(r"^[A-Z][a-z]*", name).group()
    last_name = re.search(r"[A-Z][a-z]*$", name).group()
    return player_advanced[player_advanced['PLAYER_NAME'].str.contains(first_name) & player_advanced['PLAYER_NAME'].str.contains(last_name)]
```

```python
points = [(i, lines['strength'].values[i], findMatchingLineupAdvanced(lines['lineup'].values[i])['NET_RATING'].values) for i in range(0, num_lineups)]
points = filter(lambda p: p[2].size > 0, points)
_x = map(lambda p: p[1], points)
_y = map(lambda p: p[2][0], points)
_x_np = np.array(_x).reshape((-1, 1))
_y_np = np.array(_y)
_r = linear_model.LinearRegression()
_r.fit(_x_np, _y_np)

fig, ax = plt.subplots()
scatter = ax.scatter(_x, _y)
ax.plot(_x, _r.predict(_x_np))
ax.set_xlim(min(_x) - 0.0005, max(_x) + 0.0005)
labels = [lines['lineup'].values[i].replace('+', ', ') for i in range(0, num_lineups)]
tooltip = mpld3.plugins.PointLabelTooltip(scatter, labels=labels)
mpld3.plugins.connect(fig, tooltip)
np.corrcoef(_x, _y)
```

```python
lines['NET_RTG'] = [findMatchingLineupAdvanced(lines['lineup'].values[i])['NET_RATING'].values[0] if len(findMatchingLineupAdvanced(lines['lineup'].values[i])['NET_RATING'].values) > 0 else None for i in range(0, num_lineups)]
player_net_rtgs = []
for i in range(0, num_lineups):
    players = lines['lineup'].values[i].split('+')
    net_rtgs = map(lambda p: findMatchingPlayerAdvanced(p)['NET_RATING'].values[0] if len(findMatchingPlayerAdvanced(p)['NET_RATING'].values) > 0 else None, players)
    net_rtgs = [x for x in net_rtgs if x is not None]
    player_net_rtgs.append(sum(net_rtgs) / (1. * len(net_rtgs)))
lines['PLAYER_NET_RTGS'] = player_net_rtgs
```

```python
A = np.zeros((num_lineups, num_players + 1))
b = np.zeros((num_lineups))
r = 0;
for i, row in lines.iterrows():
    players = row['lineup'].split('+')
    ids = map(lambda p: player_dict[p], players)
    for player_id in ids:
        A[r][player_id] = 1
    A[r][num_players] = row['PLAYER_NET_RTGS']
    b[r] = row['strength']
    r += 1
```

```python
t = get_model(A, b, 10)
mu = np.mean(t)
sd = np.var(t) ** 0.5
plt.hist(t, bins='auto', normed=True)
plt.plot(sorted(t), stats.norm.pdf(sorted(t), mu, sd))
plt.show()
```
